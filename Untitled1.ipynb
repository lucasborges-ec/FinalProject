```python
import numpy as np
import scipy.linalg as sc
import matplotlib.pyplot as plt
from scipy import signal
from MRPy import MRPy
```

### PROPRIEDADES DO  SERVO

### 1.0 - VELOCIDADE

```python
A = 1800    # Range do sinal quadrado (equivale a 180°)
t = 0.3     # Tempo de execução de A
v = A/t     # Velocidade fixa do servo
print (v)
```

### 2.0 AJUSTE DO SLEEPTIME

```python
t_total = 1
T_loop = 0.002277115
n = int(t_total/T_loop)
t = np.linspace(0,t_total,n*t_total)
veloc = v*t




#CASO 01
fn1 = 2
A1=400
y1 = A1*np.sin(2*np.pi*fn1*t)

#CASO 02
fn2 = 2
A2=600
y2 = A2*np.sin(2*np.pi*fn2*t)


plt.figure(5, figsize=(15,5), clear=True)
plt.ylim(0, 650);   plt.ylabel('A') 
plt.xlim(0, .5);   plt.ylabel('A') 

plt.grid(b=True, axis='x', which='major', color='b', linestyle='-')
plt.plot(t,y1);
plt.plot(t,y2);
plt.plot(t,veloc);
```
