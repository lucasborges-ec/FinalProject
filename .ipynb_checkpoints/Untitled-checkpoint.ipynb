## Decremento Logartmico

```python
t_max = 2
f = 10
n = 430

z=0.05
w=2*np.pi*f
t = np.linspace(0,t_max,n*t_max)
dec = np.exp(-(z*w*t))
plt.figure(2, figsize=(10,5), clear=True)
plt.plot(t, dec);
```
