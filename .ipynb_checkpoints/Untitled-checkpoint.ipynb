```python
# Importing Python modules required for this notebook
# (this cell must be executed with "shift+enter" before any other Python cell)

import numpy as np
import matplotlib.pyplot as plt
from   MRPy import MRPy
import numpy as np
import scipy.integrate as integrate
import gzip   as gz
import scipy as sc
```

```python

import gzip   as gz


filename='Leitura_PI3_modeoff'

with gz.open(filename+'.csv.gz', 'rb') as target:
    data =  np.genfromtxt(target, 
                          delimiter=',')
    ti   =  data[1:,0] - data[1,0]
    
a=MRPy.from_resampling(ti, data[1:,1:])

az   = MRPy(a[0], a.fs)
t    = a.t_axis()

plt.figure(4, figsize=(8, 4), clear=True)
plt.plot(t, az[0])

plt.xlim(0, az.Td);  plt.xlabel('time (s)') 
#plt.ylim(-15, 15);   plt.ylabel('a_z (m/s^2)') 

plt.grid(True) 


Saz, fs = az.periodogram()
f       = az.f_axis()

plt.figure(5, figsize=(8, 4), clear=True)
plt.plot(f, Saz[0])

plt.xlim(0, 10);   plt.xlabel('frequency (Hz)') 
plt.ylim(0,  5);   plt.ylabel('S_az (power)') 

plt.grid(True) 




```

```python


#v1 = integrate.quad(az[0],0,10)
v1=MRPy.integrate(az,band=(0,40))
t1=v1.t_axis()

u1=MRPy.integrate(v1,band=(0,40))
t2=u1.t_axis()

#print (v1)
plt.figure(6, figsize=(15, 4), clear=True)
#plt.plot(t,az[0]);
#plt.plot(t1,v1[0]);
plt.plot(t2,u1[0]);
plt.xlim(6, 50);   plt.xlabel('frequency (Hz)') 
#plt.ylim(-1,  1);   plt.ylabel('S_az (power)') 

plt.grid(True) 
```

```python
t_total = 3
n = 450
fn = 7
A = 600
sin_A=2
t = np.linspace(0,t_total,n*t_total)
y = sin_A*np.sin(2*np.pi*fn*t)

plt.figure(5, figsize=(15,5), clear=True)
plt.plot(t,y);
middle=1450
cs= (((middle+A)-(middle-A))*y-(sin_A)*((middle+A)+(middle-A)))/-(2*sin_A)
plt.figure(6, figsize=(15,5), clear=True)
plt.plot(t,cs);
```
