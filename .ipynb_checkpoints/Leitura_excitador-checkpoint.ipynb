```python
import numpy as np
import matplotlib.pyplot as plt
from   MRPy import MRPy
import numpy as np
import scipy.integrate as integrate
import gzip   as gz
import scipy as sc
```

```python
filename1='leitura_excitador/2019-11-15__06-34-28'
filename2='leitura_excitador/2019-11-15__06-49-09'
filename3='leitura_excitador/2019-11-15__06-50-31'
filename4='leitura_excitador/2019-11-15__07-46-06'
filename5='leitura_excitador/2019-11-15__08-56-22'



with gz.open(filename5+'.csv.gz', 'rb') as target:
    data =  np.genfromtxt(target, 
                          delimiter=',')
    ti   =  data[1:,0] - data[1,0]
    
a=MRPy.from_resampling(ti,10*(data[1:,1:]/8096))

az   = MRPy(a[0], a.fs)
t    = a.t_axis()

plt.figure(4, figsize=(8, 4), clear=True)
plt.plot(t, az[0])

#plt.xlim(0, az.Td);  plt.xlabel('time (s)') 
#plt.ylim(-15, 15);
plt.ylabel('a_z (m/s^2)') 

plt.grid(True) 


Saz, fs = az.periodogram()
f       = az.f_axis()

plt.figure(5, figsize=(8, 4), clear=True)
plt.plot(f, Saz[0])

plt.xlim(3, 10);   plt.xlabel('frequency (Hz)') 
#plt.ylim(0,  5);   plt.ylabel('S_az (power)') 

plt.grid(True) 
```

```python
print (az.size)
delta=t[2]-t[1]
v = np.zeros(az.size)
s = np.zeros(az.size-1)
for i in range (az.size-1):
    v[i+1] = v[i] + az[0,i]*delta    
for i in range (az.size-2):
    s[i+1] = s[i] + v[i]*delta 
```

```python
plt.figure(4, figsize=(8, 4), clear=True)
plt.plot(t, az[0]);
plt.plot(t, v);
plt.plot(t[:-1], s);

```
