```python
import numpy as np
import matplotlib.pyplot as plt
from   MRPy import MRPy
import numpy as np
import scipy.integrate as integrate
import gzip   as gz
import scipy as sc
```

```python
filename1='daniel/2019-11-18__15-15-18'
filename2='daniel/2019-11-18__15-27-47'
filename3='daniel/2019-11-18__15-28-37'
filename4='daniel/2019-11-18__16-04-12'
filename5='daniel/2019-11-18__16-03-19'
filename6='daniel/2019-11-18__16-02-42'
filename7='daniel/2019-11-18__16-04-44'
filename8='daniel/2019-11-18__17-01-26'
filename9='daniel/2019-11-18__17-02-27'
filename10='daniel/2019-11-18__17-02-58'
filename11='daniel/2019-11-18__17-01-55'

with gz.open(filename11+'.csv.gz', 'rb') as target:
    data =  np.genfromtxt(target, 
                          delimiter=',')
    ti   =  data[1:,0] - data[1,0]
    
a=MRPy.from_resampling(ti, data[1:,1:]/8096)

az   = MRPy(a[0], a.fs)
t    = a.t_axis()

plt.figure(4, figsize=(8, 4), clear=True)
plt.plot(t, az[0])

plt.xlim(0, az.Td);  plt.xlabel('time (s)') 
#plt.ylim(-15, 15);
plt.ylabel('a_z (m/s^2)') 

plt.grid(True) 


Saz, fs = az.periodogram()
f       = az.f_axis()

plt.figure(5, figsize=(8, 4), clear=True)
plt.plot(f, Saz[0])

plt.xlim(5, 9);   plt.xlabel('frequency (Hz)') 
#plt.ylim(0,  5);   plt.ylabel('S_az (power)') 

plt.grid(True) 
```
